```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("projectdb.db")
```

```python
df_overview = pd.read_sql("SELECT * FROM v_dataset_overview", conn)
print(df_overview)
```

```python
df_category = pd.read_sql("SELECT * FROM v_category_distribution", conn)
print(df_category.head(10))
```

```python
df_engagement = pd.read_sql("SELECT * FROM v_category_engagement", conn)
print(df_engagement.head(10))
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_category['category_id'], df_category['video_count'])
plt.xlabel("Category ID")
plt.ylabel("Number of Videos")
plt.title("Video Distribution by Category")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_engagement['category_id'], df_engagement['avg_engagement'])
plt.xlabel("Category ID")
plt.ylabel("Average Engagement Rate")
plt.title("Average Engagement Rate by Category")
plt.show()
```

```python
conn.close()
```
